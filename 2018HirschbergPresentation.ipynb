```python
#==============================================================================
#    IMPORTS
#==============================================================================

#-------------------------------------------------------------------------
#    Standard Libraries
#-------------------------------------------------------------------------

import matplotlib.pyplot as plt

#-------------------------------------------------------------------------
#    Local Libraries
#-------------------------------------------------------------------------

#    kCells
#--------------------------------------------------------------------

from kCells import Node
from kCells import Edge


#    Tools
#--------------------------------------------------------------------

from tools import MyLogging


#==============================================================================
#    PREPARE PLOT
#==============================================================================


with MyLogging ('pic4Hirschberg01'):
    a = 3
    n0 = Node(0,0,0)
    n1 = Node(a,0,0)
    n2 = Node(0,a,0)
    n3 = Node(0,0,a)
    nodes = [n0,n1,n2,n3]


    ax = plt.figure().add_subplot(projection='3d')
    for n in nodes:
        n.plotNode(ax)
```

```python
with MyLogging ('pic4Hirschberg02'):
    e0 = Edge(n0,n1)
    e1 = Edge(n0,n2)
    e2 = Edge(n0,n3)
    e3 = Edge(n1,n2)
    e4 = Edge(n1,n3)
    e5 = Edge(n2,n3)
    edges = [e0,e1,e2,e3,e4,e5]
    ax = plt.figure().add_subplot(projection='3d')
    for e in edges:
        e.plotEdge(ax)
```
